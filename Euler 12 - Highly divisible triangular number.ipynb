## Highly divisible triangular number

**Problem 12**

The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:


    1: 1

    3: 1,3

    6: 1,2,3,6

    10: 1,2,5,10

    15: 1,3,5,15

    21: 1,3,7,21

    28: 1,2,4,7,14,28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
def find_factors(n):
    factors = [1, n]
    i = 2
    while i * i <= n:
        if (n % i == 0):
            if(i==n/i):
                factors[int(len(factors)/2):int(len(factors)/2)]=[i]
            else:
                factors[int(len(factors)/2):int(len(factors)/2)]=[i,n/i]
            i += 1
        else:
            i+=1
    return factors
```

```python
find_factors(4)
```

```python
tri=0
i=1
number = 6
while True:
    tri +=i
    my_factors = find_factors(tri)
    print(my_factors)
    if (len(my_factors)>number):
        print(tri, len(my_factors))
        break
    i+=1
```

```python
def set_find_factors(n):
    factors = {1, n}
    i = 2
    while i * i <= n:
        if (n % i == 0):
#             print{i,n/i}
            factors = factors|{i,n/i}
            i += 1
        else:
            i+=1
    return factors
```

```python
set_find_factors(4)
```
